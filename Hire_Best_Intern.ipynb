```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load data
data = pd.read_csv('given dataset.csv')
```

```python
# Selecting only relevant columns
df_new = data.iloc[:, [1,2,3,4,5,6,10,11,12,13]]
```

```python
# Counting the number of skills and keeping in new column
df_new['Num_of_Skills'] = df_new['Other skills'].str.split(',').str.len()
score_columns = ['Performance_PG', 'Performance_UG', 'Performance_12', 'Performance_10']

# Create new columns for percentages
for column in score_columns:
    df_new[column + '_percentage'] = df_new[column].apply(lambda x: eval(x) if isinstance(x, str) else None)

# Filling Empty cells with zero
df1 = df_new.fillna(0)
```

```python
# Changing availibility answer to numerical value
df1['availability'] = df1.iloc[:, 5].apply(lambda x: 1 if isinstance(x, str) and 'yes' in x.lower() else 0)
```

```python
selected_columns = ['Python (out of 3)', 'Machine Learning (out of 3)', 'Natural Language Processing (NLP) (out of 3)', 'Deep Learning (out of 3)', 'Num_of_Skills', 'Performance_PG_percentage', 'Performance_UG_percentage', 'Performance_12_percentage', 'Performance_10_percentage', 'availability']

# Creating a new dataframe with the selected columns
new_df = df1[selected_columns]
```

```python
column_rename_dict = {
    'Python (out of 3)': 'Python_Score',
    'Machine Learning (out of 3)': 'ML_Score',
    'Natural Language Processing (NLP) (out of 3)': 'NLP_Score',
    'Deep Learning (out of 3)': 'DL_Score',
    'Performance_PG_percentage': 'PG_Score',
    'Performance_UG_percentage': 'UG_Score',
    'Performance_12_percentage': '12_Score',
    'Performance_10_percentage': '10_Score'
}

# Rename the columns using the dictionary
new_df = new_df.rename(columns=column_rename_dict)
```

```python
# Step 2 : Normalize the data
columns_to_normalize = ['Python_Score', 'ML_Score', 'NLP_Score', 'DL_Score', 'Num_of_Skills']

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Normalize the selected columns
new_df[columns_to_normalize] = scaler.fit_transform(new_df[columns_to_normalize])
new_df.head()
```

```python
# Upon checking found two candidates score was beyond 1 fixing that

new_df.loc[1078, 'PG_Score'] = 0.805
new_df.loc[200, 'UG_Score'] = 1
```

```python

# Assumining the weights
criteria_weights = {
    'Python_Score': 0.1,
    'ML_Score': 0.1,
    'NLP_Score': 0.1,
    'DL_Score': 0.1,
    'Num_of_Skills': 0.05,
    'UG_Score': 0.1,
    'PG_Score': 0.1,
    '12_Score': 0.1,
    '10_Score': 0.1,
    'availability': 0.1
}

# Step 3: Calculate scores
new_df['score'] = 0  # Create a new column to store the scores
for criterion, weight in criteria_weights.items():
    new_df['score'] += new_df[criterion] * weight

# Step 4: Rank candidates
new_df = new_df.sort_values(by='score', ascending=False)

# Step 5: Select the best candidate
best_candidate = new_df.iloc[0]

new_df.head()
```
